## 1.2 — Conditions & Branching

Conditionals appear everywhere: thresholding predictions, flagging outliers, filtering data.

```python
# ── Conditions Example ────────────────────────────────────────────────────
def classify_bmi(bmi: float) -> str:
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25.0:
        return "Normal"
    elif bmi < 30.0:
        return "Overweight"
    else:
        return "Obese"

test_bmis = [16.2, 22.4, 27.8, 35.1]
for b in test_bmis:
    print(f"BMI {b:5.1f} → {classify_bmi(b)}")
```

## 1.4 — List Comprehensions

One of Python's most powerful one-liners. In AI you use them constantly to transform feature columns, filter rows, and build label arrays.

```python
# ── List Comprehension Examples ───────────────────────────────────────────
ages = [52, 53, 70, 61, 62, 58, 45, 38, 67, 72]

# 1. Convert ages to "Senior" / "Adult"
age_group = ["Senior" if a >= 60 else "Adult" for a in ages]
print("Age groups:", age_group)

# 2. Normalise ages to [0, 1]  (min-max scaling — what sklearn does internally)
min_a, max_a = min(ages), max(ages)
ages_norm = [(a - min_a) / (max_a - min_a) for a in ages]
print("Normalised:", [round(v, 3) for v in ages_norm])

# 3. Filter: keep only seniors
seniors = [a for a in ages if a >= 60]
print("Seniors only:", seniors)
```

## 1.5 — Containers: List, Tuple, Dict, Set

Each container has a role in ML pipelines:

| Container | Mutable | Ordered | Use in AI |
|-----------|---------|---------|-----------|
| `list`    | ✅ | ✅ | Store feature arrays, batch samples |
| `tuple`   | ❌ | ✅ | Fixed config (input_shape, random_state) |
| `dict`    | ✅ | ✅ (3.7+) | Hyperparameters, label maps, metrics |
| `set`     | ✅ | ❌ | Unique classes, duplicate detection |

```python
# ── Containers in Action ──────────────────────────────────────────────────

# dict → hyperparameter config (you'll see this pattern in every sklearn/pytorch project)
config = {
    "model"         : "RandomForest",
    "n_estimators"  : 100,
    "max_depth"     : 5,
    "random_state"  : 42,
    "learning_rate" : 0.01,
}
print("Config:", config)
print("n_estimators:", config["n_estimators"])

# tuple → immutable model input shape
input_shape = (224, 224, 3)    # height, width, channels
print("\nInput shape:", input_shape)

# set → find unique class labels in a predictions list
predictions = [0, 1, 0, 1, 1, 0, 2, 2, 1, 0]
unique_classes = set(predictions)
print("\nUnique classes:", unique_classes)

# Iterating a dict — common when logging metrics
metrics = {"accuracy": 0.92, "precision": 0.89, "recall": 0.94, "f1": 0.91}
for metric, value in metrics.items():
    print(f"  {metric:12s}: {value:.2f}")
```

## 1.6 — Functions & Lambda

Reusable functions keep your AI notebooks clean. `lambda` is used in `map()`, `apply()`, and sorting.

```python
# ── Functions ─────────────────────────────────────────────────────────────

def min_max_scale(values: list) -> list:
    """Scale a list of numbers to [0, 1]."""
    lo, hi = min(values), max(values)
    return [(v - lo) / (hi - lo) for v in values]

def one_hot_encode(labels: list) -> dict:
    """Simple one-hot encoder returning a dict of binary lists."""
    classes = sorted(set(labels))
    return {cls: [1 if l == cls else 0 for l in labels] for cls in classes}

# Test
bp = [120, 140, 130, 160, 110]
print("Scaled BP:", [round(v, 3) for v in min_max_scale(bp)])

labels = ["cat", "dog", "cat", "bird", "dog"]
print("\nOne-hot:")
for cls, vec in one_hot_encode(labels).items():
    print(f"  {cls}: {vec}")

# Lambda — quick inline transforms
to_binary = lambda x: 1 if x > 0 else 0
raw_scores = [-0.5, 2.1, 0.0, 1.3, -1.0]
print("\nBinary outputs:", list(map(to_binary, raw_scores)))
```

## 2.4 — Handling Missing Values & Data Cleaning

Real-world data is messy. This is how you clean it for scikit-learn.

```python
# ── Simulate a dataset with missing values ────────────────────────────────
np.random.seed(42)
df_dirty = df.copy()

# Randomly inject NaN into 5% of cells
for col in ["chol", "trestbps", "thalach"]:
    mask = np.random.random(len(df_dirty)) < 0.05
    df_dirty.loc[mask, col] = np.nan

print("Missing values introduced:")
print(df_dirty[["chol", "trestbps", "thalach"]].isnull().sum())
```

```python
# ── Strategy 1: Drop rows with ANY missing value ──────────────────────────
df_dropped = df_dirty.dropna()
print(f"Rows after dropna(): {len(df_dropped)}  (original: {len(df_dirty)})")

# ── Strategy 2: Fill with median (robust to outliers — preferred in medical data)
df_filled = df_dirty.copy()
for col in ["chol", "trestbps", "thalach"]:
    median_val = df_filled[col].median()
    df_filled[col] = df_filled[col].fillna(median_val)
    print(f"  Filled '{col}' missing values with median = {median_val:.1f}")

print(f"\nMissing after fill: {df_filled.isnull().sum().sum()}")
```

## 2.5 — Feature Engineering

Creating new features from existing ones often boosts model performance more than tuning hyperparameters.

```python
# ── Feature Engineering Examples ──────────────────────────────────────────
df_feat = df.copy()

# 1. Age group (ordinal encoding)
df_feat["age_group"] = pd.cut(
    df_feat["age"],
    bins=[0, 40, 55, 65, 100],
    labels=["Young", "Middle", "Senior", "Elderly"]
)

# 2. Cholesterol category
df_feat["chol_cat"] = df_feat["chol"].apply(
    lambda x: "High" if x > 240 else ("Borderline" if x > 200 else "Normal")
)

# 3. Heart rate reserve (max HR − achieved HR; rough fatigue indicator)
df_feat["hr_reserve"] = 220 - df_feat["age"] - df_feat["thalach"]

print(df_feat[["age", "age_group", "chol", "chol_cat", "thalach", "hr_reserve"]].head(8))
```

## 2.6 — Encoding Categorical Variables for scikit-learn

scikit-learn models require **all-numeric** input. Here's how to encode categoricals.

```python
# ── Label Encoding ────────────────────────────────────────────────────────
from sklearn.preprocessing import LabelEncoder, StandardScaler

df_enc = df_feat[["age", "chol", "thalach", "sex", "target"]].copy()

# Confirm dtypes are all numeric before feeding to sklearn
print("Dtypes before encoding:")
print(df_enc.dtypes)
print("\n✅ All columns are numeric — ready for sklearn!")
```

```python
# ── StandardScaler — zero mean, unit variance ─────────────────────────────
num_cols = ["age", "chol", "thalach"]

scaler   = StandardScaler()
X_scaled = scaler.fit_transform(df_enc[num_cols])

df_scaled = pd.DataFrame(X_scaled, columns=[c + "_scaled" for c in num_cols])
print("Scaled features (first 5 rows):")
print(df_scaled.head())
print(f"\nMean (should ≈ 0): {df_scaled.mean().round(4).to_dict()}")
print(f"Std  (should ≈ 1): {df_scaled.std().round(4).to_dict()}")
```
